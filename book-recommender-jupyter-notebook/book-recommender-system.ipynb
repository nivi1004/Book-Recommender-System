## Collaborative Filtering

```python
ratings_with_name
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
active_users = x[x].index
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python

final_ratings
```

```python
final_ratings.head()
```

```python
# Ensure that the title used in the pivot table is the lowercase version
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
famous_books = filtered_rating.groupby('lowercase_title').count()['Book-Rating'] >= 50
famous_books = famous_books[famous_books].index

final_ratings = filtered_rating[filtered_rating['lowercase_title'].isin(famous_books)]

# Create the pivot table using the lowercase title
pt = final_ratings.pivot_table(index='lowercase_title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score
```

```python
cosine_similarity(pt).shape
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
    #index fetch
    book_name = book_name.lower()

    # Index fetch using lowercase title
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        # Use the original case title for display
        temp_df = books[books['lowercase_title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Author')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Notebook')
```

```python
recommend('The Da Vinci Code')
```
